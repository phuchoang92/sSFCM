# **PreProcess**

## Load Data

```python
import pandas as pd
data = pd.read_csv('resources/iris.data', header = None)
data = data.select_dtypes(include=["float64", "int64"])
data
```

```python
import numpy as np
X = np.array(data)
n = X.shape[0]
p = X.shape[1]
print(X.shape)
#Maxtrix X size= n*7
X
```

## Generate U_ngang , V

```python
from random import random
c=3   #value of c 
U_ngang = np.random.rand(n,c)
sum_i = np.zeros(n) 
# Doan nay lam tong u_ki <= 1 , i = 1,c 
for i in range(n):
    U_ngang[i] = U_ngang[i]/sum(U_ngang[i])*random()
    sum_i[i] = sum(U_ngang[i])
#Maxtrix U_ngang
U_ngang
```

```python
#Maxtrix V sinh ngau nhien
V = np.random.rand(c,p)
V_truoc = V
V
```

## Function 4 ,6

```python
# Tinh d_ki (trong cong thuc 6)
def d_ki(k, i):
    kq = sum(  pow(X[k] - V[i], 2) )
    return kq     
d_ki(0,0)
```

```python
#Cong thuc so 6

U = np.zeros((n,c))  # Maxtrix U


def cong_thuc_6(m):
    
    Z = np.zeros((n,c))
    #tu day suy ra U = U_ngang + Z ta tinh Z

    D = np.zeros((n,c))
    for k in range(n):
        for i in range(c):
            D[k][i] = pow(d_ki(k,i),1/(1-m))
            D[k] = D[k]/sum(D[k])    
     

    #Z = (1/d_ki)^1/m-1z = np.array (i = 1 ,c)
    for i in range(n) :
        Z[i] = (1 - sum_i[i])*D[i]
        
    return U_ngang+Z  # tuc la U  
```

```python
def cong_thuc_4(m):
    global V, V_truoc
    V_truoc = V
    print("V_truoc:")
    print(V_truoc)
    V_temp = np.zeros((c,p))
    U_temp = abs(U - U_ngang)
    U_temp = pow(U_temp,1/(m-1))
    mau_so_array= sum(U_temp.T)
    for i in range(c):
       tu_so = np.zeros(p)
       for k in range(n):
           tu_so +=U_temp[k][i] * X[k] 
       V_temp[i] = tu_so/ sum(mau_so_array)
    print("V_sau:")
    print(V_temp)
    return V_temp
```

```python
def thuat_toan(epsilon,m):
    global U,V
    Epsilon = np.zeros((c,p)) + epsilon
    i=1
    while True:
        print("-----------------------\nChay lan "+ str(i)+"\n")
        i+=1
        U= cong_thuc_6(m)
        V= cong_thuc_4(m)
        delta_V = abs(V- V_truoc)
        ktra = np.less_equal(delta_V, Epsilon)
        if (np.all(ktra)):
            break
    print("\n------------------------------------\nKetqua\n")
    print(V)
    return
```
