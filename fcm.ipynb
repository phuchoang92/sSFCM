# **Dataset Preparation & Preprocessing**

## Load Data

```python
import pandas as pd
data = pd.read_csv('resources/iris.data', header = None)
data = data.select_dtypes(include=["float64", "int64"])
data
```

```python
import numpy as np
X = np.array(data)
n = X.shape[0]
p = X.shape[1]
print(X.shape)
X
```

## Generate U_ngang , V

```python
c = 0
def set_c(input):
    global c
    c = input
```

```python
from random import random
import random

def generate_U_ngang(k1):
    global U_ngang
    global index_x_giamsat
    U_ngang = np.zeros((n,c))
    
    index_x_giamsat = random.sample(range(n),k1)# CHọn k1 vector x từ n vector x ban đầu  
    index_x_giamsat = np.sort(index_x_giamsat)
    # print(index_x_giamsat)

    for i in range(k1):
        c1= random.randint(0,c-1)                    #Chọn vị trí cụm giám sát
        U_ngang[index_x_giamsat[i]][c1] = 0.5     
    global sum_i 
    sum_i= sum(U_ngang.T)
# #Maxtrix U_ngang
# for i in range(n):
#     print(str(i) +" : "+ str(U_ngang[i]))
```

```python
#Maxtrix V sinh ngau nhien

def generate_V():
    global V, V_truoc
    V = np.random.rand(c,p) 
    V_truoc = V
```

## Function 4 ,6

```python
# Tinh d_ki (trong cong thuc 6)
def d_ki(k, i):
    kq = sum(  pow(X[k] - V[i], 2) )
    return kq     
```

```python
#Cong thuc so 6

U = np.zeros((n,c))  # Maxtrix U


def cong_thuc_6(m):
    
    Z = np.zeros((n,c))
    #tu day suy ra U = U_ngang + Z ta tinh Z

    D = np.zeros((n,c))
    for k in range(n):
        for i in range(c):
            D[k][i] = pow(d_ki(k,i),1/(1-m))    
        D[k] = D[k]/sum(D[k])    
      
    #Z = (1/d_ki)^1/m-1z = np.array (i = 1 ,c)
    for i in range(n) :
        Z[i] = (1 - sum_i[i] )*D[i]
        
    return U_ngang+Z  # tuc la U  
```

```python
def cong_thuc_4(m):
    global V, V_truoc
    V_truoc = V
    # print("V_truoc:")
    # print(V_truoc)
    V_temp = np.zeros((c,p))
    U_temp = abs(U - U_ngang)
    U_temp = pow(U_temp,m)
    for i in range(c):
        tu_so = np.zeros(p)
        for k in range(n):
           tu_so +=U_temp[k][i] * X[k] 
        V_temp[i] = tu_so/ sum((U_temp.T)[i])
    # print("V_sau:")
    # print(V_temp)
    return V_temp
```

# **Đánh giá kết quả**

```python
def count(c):
    global count_class_cluster
    count_class_cluster = np.zeros((2,c), dtype="int64")
    for k in range(n):
        if k in index_x_giamsat :
            k_class = 0
        else :
            k_class = 1
        max_u = np.amax(U[k])
        index_max = np.where(U[k]== max_u )
    
        count_class_cluster[k_class][index_max] +=1   
     
```

```python


num_class = 2


def external_validity(c):
    a1 = b1 = c1 = d1= 0
    for i in range(num_class):
        a1+= sum ( (count_class_cluster[i]-1)*count_class_cluster[i]/2 )
        b1+= (pow(sum(count_class_cluster[i]),2) - sum(pow(count_class_cluster[i],2)) )/2
    for i in range(c):
        c1+= count_class_cluster[0][i]*count_class_cluster[1][i]
    for j in range(c):
        if i != j :
            d1+=count_class_cluster[0][i]*count_class_cluster[1][j] 
            
    # print(a1)
    # print(b1)
    # print(c1)
    # print(d1)
    # Rand int
    w1 =  (a1 + d1)/ (a1 +b1 + c1 +d1)
    # Adjusted Rand Index
    M= a1 + b1 +c1 +d1
    w2 = (a1 - (a1+c1)*(a1+b1)/M) /( (2*a1+b1+c1)/2 - (a1+c1)*(a1+b1)/M )
    #Jaccard Coefficient
    w3 = a1 / (a1+b1+c1)
    print ("Rant Index :          w1 = " + str(w1))
    print ("Adjusted Rand Index : w2 = " + str(w2))
    print ("Jaccard Coefficient : w3 = " + str(w3))
    
```

# **Thuật toán**

```python
def thuat_toan(epsilon,m,c,k1):
    set_c(c)
    
    generate_U_ngang(k1)
    generate_V()

    global U,V,count_class_cluster
    Epsilon = np.zeros((c,p)) + epsilon
    i=1
    while True:
        # print("-----------------------\nChay lan "+ str(i)+"\n")
        i+=1
        U= cong_thuc_6(m)
        V= cong_thuc_4(m)
        delta_V = abs(V- V_truoc)
        ktra = np.less_equal(delta_V, Epsilon)
        if (np.all(ktra)):
            break
    print("\n------------------------------------\nKetqua\nV:")
    print(V)
    count(c)
    count_X = pd.DataFrame(count_class_cluster)
    external_validity(c)
    return
```
