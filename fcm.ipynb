```python
import pandas as pd
data = pd.read_csv('abalone.data', header = None, usecols= [1,2,3,4,5,6,7])
data
```

```python
import numpy as np
X = np.array(data)
n = int(X.size/7)
print(X.shape)
#Maxtrix X size= n*7
X
```

```python
from random import random
c=3   #value of c 
U_ngang = np.random.rand(n,c)
sum_i = np.zeros(n) 
# Doan nay lam tong u_ki <= 1 , i = 1,c 
for i in range(n):
    U_ngang[i] = U_ngang[i]/sum(U_ngang[i])*random()
    sum_i[i] = sum(U_ngang[i])
#Maxtrix U_ngang
U_ngang
```

```python
V = np.random.rand(c,7)
#Maxtrix V sinh ngau nhien
V
```

```python
# Tinh d_ki (trong cong thuc 6)
def d_ki(k, i):
    kq = sum(  pow(X[k] - V[i], 2) )
    return kq     
d_ki(0,0)
```

```python
#Cong thuc so 6

U = np.zeros((n,c))  # Maxtrix U


def cong_thuc_6(m):
    
    Z = np.zeros((n,c))
    #tu day suy ra U = U_ngang + Z ta tinh Z

    D = np.zeros((n,c))
    for k in range(n):
        for i in range(c):
            D[k][i] = pow(d_ki(k,i),1/(1-m))
            D[k] = D[k]/sum(D[k])    
     

    #Z = (1/d_ki)^1/m-1z = np.array (i = 1 ,c)
    for i in range(n) :
        Z[i] = (1 - sum_i[i])*D[i]
        
    return U_ngang+Z  # tuc la U  
```

```python
def cong_thuc_4(m):
    V_temp = np.zeros((c,7))
    U_temp = abs(U - U_ngang)
    U_temp = pow(U_temp,1/(m-1))
    mau_so_array= sum(U_temp.T)
    for i in range(c):
       tu_so = np.zeros(7)
       for k in range(n):
           tu_so +=U_temp[k][i] * X[k] 
       V_temp[i] = tu_so/ sum(mau_so_array)
    return V_temp
```
